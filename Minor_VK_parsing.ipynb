```python
items = []
for i in range(0, 7):
    res = api.wall.getComments(owner_id = -29534144, post_id = 12925514, need_likes = 1, offset = 100*i, sort = 'asc', preview_length = 0, count = 100, extended = 1, v = 5.33)
    items_res = res['items']
    items.extend(items_res)
    sleep(0.5)
```

```python
full_list = []
for item in items:
    try:
        L = [item['from_id'], item['id'], item['text'], item['date'],
            item['likes']['count'], item['reply_to_user']]
    except:
        L = [item['from_id'], item['id'], item['text'], item['date'],
            item['likes']['count'], np.nan]
    full_list.append(L)
```

```python
df = pd.DataFrame(full_list)
```

```python
df.columns = ['from_id', 'id', 'text', 'date', 'likes', 'reply_to_user']
df.head(10)
```

```python
from datetime import datetime

def date_norm(date):
    d = datetime.fromtimestamp(date) # timestamp - UNIX-–≤—Ä–µ–º—è –≤ –≤–∏–¥–µ —Å—Ç—Ä–æ–∫–∏
    str_d = d.strftime("%Y-%m-%d %H:%M:%S") # %Y-%m-%d %H:%M:%S - –≥–æ–¥-–º–µ—Å—è—Ü-–¥–µ–Ω—å, —á–∞—Å—ã:–º–∏–Ω:—Å–µ–∫
    return str_d

df['date'] = df['date'].apply(date_norm)
```
